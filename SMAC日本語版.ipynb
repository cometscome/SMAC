初期化ルーチン

グリーン関数$G(\tau)$をファイルから読み出す。  
サンプルのグリーン関数は、IQISTによって計算されている。  
ベーテ格子のハーフフィリング、β=100でU=2の時の動的平均場理論による結果を使用。

```julia

ファイル名 = "./solver.green.dat"
周波数の数 = 256 
行の数 = countlines(open( ファイル名, "r" ))
行の塊 = readlines(ファイル名)
最後の行 = 行の塊[行の数-2]
数値化された最後の行 = split(最後の行)
軌道の数 = parse(Int64,数値化された最後の行[1])
虚時間の数 = parse(Int64,数値化された最後の行[2])
β = parse(Float64,数値化された最後の行[3])
最大周波数 = 5.0
println("周波数の刻み数: ",周波数の数,"\n","虚時間の数: ",虚時間の数)
println("軌道の数: ",軌道の数)
println("逆温度: ",β)


cnt = 0
虚時間配列 = zeros(Float64,虚時間の数,軌道の数)
Gtau = zeros(Float64,虚時間の数,軌道の数,2)
for orbital in 1:軌道の数
    for itime in 1:虚時間の数
        cnt += 1
        u = split(行の塊[cnt])
        #println(u[3])
        #println(u[2])
        虚時間配列[itime,orbital] = parse(Float64,u[3])
        Gtau[itime,orbital,1] = parse(Float64,u[4])
        Gtau[itime,orbital,2] = parse(Float64,u[5])
        #println(temperatures[itemperature,orbital])
        #println(cnt)
    end
    cnt += 2
end

周波数配列 = zeros(Float64,周波数の数)
dω = 2最大周波数/(周波数の数-1)
for i in 1:周波数の数
    周波数配列[i] = (i-1)*dω - 最大周波数
end


plot(虚時間配列[:,1],-Gtau[:,1,1],label="original",title="Imaginary-time Green's function")
```

# Main routine

```julia
include("./smac日本語版.jl")
using Smac
vec_Gout = zeros(Float64,虚時間の数,1)
xout = zeros(Float64,周波数の数,1)


@time (xout,vec_Gout) = Smac.smac_main!(虚時間の数,周波数の数,1,最大周波数,β,-Gtau)
```

# Results

```julia
plot(虚時間配列[:,1],[-Gtau[:,1,1],vec_Gout[:,1]],label=["Original","SMAC"],title="Imaginary-time Green's function")
```

```julia
plot(周波数配列[:],xout[:,1]/dω,label="1",title="Spectral functions", marker=:circle)
```
